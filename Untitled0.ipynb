```python
#importing libraries necessary
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as datetime

# Load the dataset
file_path = "Test Data.xlsx"  # Update with the correct file path
data = pd.read_excel('/practice data.xlsx')
data.head()
data.dropna(inplace=True)

```

# New Section

```python
# Convert start_date to datetime format
data['start_date'] = pd.to_datetime(data['start_date'])

data['start_month'] = data['start_date'].dt.to_period('M')
data.head()
```

```python

grouping = data.groupby('company_id')['start_month']
data['cohort_day']= grouping.transform( 'min')


data.head()
```

```python
data['cohort_index'] = (data['start_month'] - data['cohort']).apply(lambda x: x.n)

# Step 3: Build the cohort matrix (number of companies per cohort and month)
cohort_pivot = data.pivot_table(index='cohort', columns='cohort_index', values='company_id', aggfunc=pd.Series.nunique)
```

```python
def percentage_sign (x, pos):
  return f'{x:.0f}%'
```

```python
# Convert to percentage retention rates
cohort_size = cohort_pivot.iloc[:, 0]
retention_matrix = cohort_pivot.divide(cohort_size, axis=0)*100

# Step 4: Customizing the heatmap
plt.figure(figsize=(22, 9))
sns.heatmap(retention_matrix, annot=True, fmt=".0f", cmap="Blues", linewidths=0.5, linecolor='gray')

# Customizing the heatmap aesthetics
plt.title("Customized Cohort Analysis - Company Retention Over Time", fontsize=14, fontweight='bold')
plt.xlabel("Months Since First Campaign", fontsize=12)
plt.ylabel("Cohort (First Campaign Month)", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10, rotation=0)
plt.show()
```
